```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

data = pd.DataFrame({
    "Umur": [20, 30, 40, 50, 60],
    "Pendapatan": [ 3_500_000, 4_500_000, 6_000_000, 9_000_000, 12_000_000]
})

minmax = MinMaxScaler()
data_minmax = minmax.fit_transform(data)
print(data_minmax)
standard = StandardScaler()
data_standard = standard.fit_transform(data)
data_standard
```
