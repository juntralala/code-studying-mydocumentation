## Scatter Diagram

```python
import matplotlib.pyplot as plt
x = ["Buku", "Pulpen", "Penggaris", "Pulpen"]
y = [120, 20, 50, 52]

plt.scatter(x=x, y=y)

# Agar nilai sumbu x dan y mulai dari 0
plt.axhline(0, linewidth=0)
plt.axvline(0, linewidth=0)

plt.show()
```

```python
# Sedikit lebih advance, tanpa axes
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 12))

data = pd.DataFrame({
    "Nama": ["Buku", "Pensil", "Penghapus"],
    "Jumlah": [5, 6, 9]
})

"""
panjang colors harus lebih atau sama dengan jumlah titik. Jika tidak, maka error.
Nilai maks dan minimum warna adalah bebas, karna nilai paling kecil 1 dan maks 3, maka warna paling biru adalah 1 dan warna paling merah adalah 3 (pada cmpa="coolwarm")
"""
colors = [1, 2, 3]
# colors = "red" # pakainama warna langsung, cmap diabaikan

# cmap => tema warna yg digunakan
plt.scatter(data["Nama"], data["Jumlah"], c=colors, cmap="coolwarm")

# Menambah teks diatas salah satu titik
plt.annotate(
    text="Buku:5",
    xy=(data['Nama'][0], data["Jumlah"][0]), # kordinat titik
    xytext=(0, 8), # x=0, y=8 dari offset points: teks sedikit keatas
    textcoords="offset points",
    ha="center",
    color="dimgray"
)

# plt.axhline(0, linewidth=0) # agar sumbu y mulai 0

plt.colorbar(label="Nilai")
plt.title("Nyoba bikin diagram Scatter")
plt.ylabel("Nama Produk")
plt.xlabel("Jumlah Terjual")
plt.grid(True, alpha=0.4)

# plt.tight_layout()
plt.show()
```
