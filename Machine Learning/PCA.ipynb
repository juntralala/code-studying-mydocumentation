```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

# [usia, tinggi(cm), berat(kg), sistolik, diastolik, BMI]
data = [
    [25, 165, 55, 110, 70, 20.2],
    [30, 170, 68, 118, 75, 23.5],
    [35, 175, 80, 125, 82, 26.1],
    [40, 160, 72, 130, 85, 28.1],
    [45, 180, 90, 135, 88, 27.8],
    [50, 155, 60, 140, 90, 24.9],
    [55, 168, 75, 145, 92, 26.6],
    [60, 172, 85, 150, 95, 28.7],
    [65, 158, 65, 155, 98, 26.0],
    [70, 163, 70, 160, 100, 26.3],
]

data = np.array(data)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

print("\nData setelah normalisasi:\n", data_scaled.round(3))

pca = PCA(n_components=6)
transformed = pca.fit_transform(data_scaled)

print("\n--- Hasil PCA ---")
print("Shape setelah PCA:", transformed.shape)
print("\nTransformed data (5 baris pertama):\n", transformed[:5].round(4))
print("\nExplained variance ratio:", pca.explained_variance_ratio_.round(4))
print("Total variance explained: {:.2f}%".format(
    pca.explained_variance_ratio_.sum() * 100
))

print(pca.components_)
print(pca.singular_values_)
print(pca._parameter_constraints)
```

```python
import matplotlib.pyplot as plt

plt.plot(data_scaled[:,0], c="purple")
plt.plot(data_scaled[:,1], c="red")
plt.plot(data_scaled[:,2], c="yellow")
plt.plot(data_scaled[:,3], c="orange")
plt.plot(data_scaled[:,4], c="green")
plt.plot(data_scaled[:,5], c="blue")
plt.show()
```

```python
plt.plot(transformed[:,0], c="red")
plt.plot(transformed[:,1], c="blue")
plt.plot(transformed[:,2], c="yellow")
plt.plot(transformed[:,3], c="orange")
plt.show()
```
