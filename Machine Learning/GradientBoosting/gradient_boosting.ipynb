```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import FeatureUnion
```

```python
df = pd.DataFrame({
    'jenis_kelamin': ['Laki-laki', 'Perempuan', 'Laki-laki', 'Perempuan', 'Laki-laki',
                      'Perempuan', 'Laki-laki', 'Perempuan', 'Laki-laki', 'Perempuan',
                      'Laki-laki', 'Perempuan', 'Laki-laki', 'Perempuan', 'Laki-laki'],
    'tinggi_cm':     [165, 153, 168, 150, 163,
                      156, 170, 148, 167, 154,
                      162, 151, 169, 152, 164],
    'berat_kg':      [60,  48,  65,  45,  58,
                      50,  68,  43,  62,  47,
                      55,  46,  63,  49,  57]
})

x = df.drop(columns='jenis_kelamin')
y = df['jenis_kelamin']


model = GradientBoostingClassifier()
model.fit(x, y)
model.predict([[175, 70]])
```

```python
model.classes_
```

```python

model.predict_proba([[175, 70]])
```

```python
# Finetuning
params = {
    "n_estimators": [100, 125, 150, 175, 200],
    "max_depth": [3, 4, 5],
    "min_samples_leaf": [1, 2, 3],
    "min_samples_split": [2, 3],
    "learning_rate": [0.1, 0.01, 0.001]
}

model = GradientBoostingClassifier(random_state=42, criterion="squared_error")
model_tuned = RandomizedSearchCV(estimator=model,
                           param_distributions=params,
                           n_iter=30,
                           cv=3,
                           random_state=42,
                           verbose=1,
                           n_jobs=-1)
model_tuned.fit(x, y)
print(f"Score: {model_tuned.best_score_}")
model_tuned.best_estimator_
```
